```python
from matplotlib.colors import Normalize
import os
import json
import torch
import numpy as np

def calculate_attention_values(attention_matrix, input_tokens, output_tokens, image_token_start_idx, image_token_end_idx):
    # Calculate the average image tokens for each output token before normalizing
    average_attention_scores = []

    # calculate rationale's attention scores
    for i, output_token in enumerate(output_tokens):
        output_relevancy = attention_matrix[len(input_tokens) - 1 + i, :]
        image_tokens = output_relevancy[image_token_start_idx:image_token_end_idx]
        text_tokens = output_relevancy[image_token_end_idx + 1:len(input_tokens)-3]

        # calculate rationale token
        answer_length = 15
        average_rationale_tokens = None
        if len(output_tokens) > (answer_length*2):
            rationale_length = (len(output_tokens) - answer_length)
            if i >= rationale_length:
                rationale_tokens = output_relevancy[len(input_tokens) - 1: len(input_tokens) - 1 + rationale_length]
                average_rationale_tokens = np.mean(rationale_tokens)
    
        average_text_tokens = np.mean(text_tokens)
        average_image_tokens = np.mean(image_tokens)
        
        average_attention_score = {
            "id": i,
            "output_token": output_token,
            "average_text_tokens": average_text_tokens,
            "average_image_tokens": average_image_tokens,
            "average_image_to_text_ratio": average_image_tokens / average_text_tokens,
            "text_tokens": text_tokens,
            "image_tokens": image_tokens
        }
        if average_rationale_tokens is not None:
            average_attention_score["average_rationale_tokens"] = average_rationale_tokens
            average_attention_score["average_rationale_to_image_ratio"] = average_rationale_tokens / average_image_tokens 
            average_attention_score["average_rationale_to_text_ratio"] = average_rationale_tokens / average_text_tokens 

        average_attention_scores.append(average_attention_score)

    return average_attention_scores 

def calculate_image_text_ratio(attention_scores):
    ratio_list = []
    average_image_list = []
    average_rationale_to_image_list = []
    average_rationale_to_text_list = []
    text_list = []
    for output_token in attention_scores:
        ratio_list.append(output_token["average_image_to_text_ratio"])
        average_image_list.append(output_token["average_image_tokens"])
        if "average_rationale_to_image_ratio" in output_token and output_token["average_rationale_to_image_ratio"] is not None:
            average_rationale_to_image_list.append(output_token["average_rationale_to_image_ratio"])
        if "average_rationale_to_text_ratio" in output_token and output_token["average_rationale_to_text_ratio"] is not None:
            average_rationale_to_text_list.append(output_token["average_rationale_to_text_ratio"])
        text_list.append(output_token["text_tokens"])

    
    return np.mean(np.array(ratio_list)), np.mean(np.array(average_image_list)), np.mean(np.array(text_list), axis=0), np.mean(np.array(average_rationale_to_image_list)), np.mean(np.array(average_rationale_to_text_list))

def process_data(file_names, base):
  processed_data = {key: {'items': []} for key in file_names}
  for key, file_name in file_names.items():
    with open(os.path.join(base, file_name, "results.json"), "r") as file:
      results = json.load(file)

    for i, result in enumerate(results):
      if result["predicted_answer"] == "":
          continue;
          
      average_attention_matrix = torch.load(os.path.join(base, file_name, "attention_weights", f"{i}.pt"), map_location=torch.device('cpu'))
      attention_file = os.path.join(base, file_name, "attention", f"{i}.json")
      with open(attention_file, "r") as file:
          attention_description = json.load(file)

      # filter tokens
      output_tokens = [token.replace('Ġ', '').replace('Ċ', '').replace('Ģ','').replace('Ï','') for token in attention_description["output_tokens"]]
      input_tokens = [token.replace('Ġ', '').replace('Ċ', '').replace('Ģ','').replace('Ï','') for token in attention_description["input_tokens"]]
      image_token_indices = [i for i, token in enumerate(input_tokens) if token.startswith("image_")]
      image_token_start_idx = image_token_indices[0]
      image_token_end_idx = image_token_indices[-1] + 1
      main_input_text_tokens = input_tokens[image_token_end_idx + 1:len(input_tokens)-3]

      # calculate attention values for each token!
      attention_results = calculate_attention_values(average_attention_matrix, input_tokens, output_tokens, image_token_start_idx, image_token_end_idx)
      image_text_ratio, average_image, average_text_tokens, average_rationale_to_image_ratio, average_rationale_to_text_ratio = calculate_image_text_ratio(attention_results)

      # an item, is a specific problem
      # store in processed data structure
      item_data = {
          "idx": i,
          'attention_results': attention_results,
          'input_tokens': input_tokens,
          'output_tokens': output_tokens,
          'image_token_start_idx': image_token_start_idx,
          'image_token_end_idx': image_token_end_idx,
          'average_image_to_text_ratio': image_text_ratio,
          'main_input_text_tokens': main_input_text_tokens,
          'average_image': average_image,
          'average_text_tokens': average_text_tokens,
          'average_rationale_to_image_ratio': average_rationale_to_image_ratio,
          'average_rationale_to_text_ratio': average_rationale_to_text_ratio,
          'predicted_answer': result["predicted_answer"],
          'problem': {
              "problem_text": result["problem_text"],
              "choices": result["choices"],
              "answer": result["answer"],
              "image": result["image_id"] + ".png",
              "predicted_answer": result["predicted_answer"],
              "extracted_answer": result["extracted_answer"],
              "problem_type_graph": result["problem_type_graph"],
              "problem_type_goal": result["problem_type_goal"],
          }
      }
      processed_data[key]['items'].append(item_data)

    # this is for ALL problems
    average_images = np.mean([item['average_image'] for item in processed_data[key]['items']])
    processed_data[key]['average_image'] = average_images
      
    average_ratios = np.mean([item['average_image_to_text_ratio'] for item in processed_data[key]['items']])
    processed_data[key]['average_image_to_text_ratio'] = average_ratios

    average_ratios_2 = np.nanmean([item['average_rationale_to_image_ratio'] for item in processed_data[key]['items'] if "average_rationale_to_image_ratio" in item])
    processed_data[key]['average_rationale_to_image_ratio'] = average_ratios_2

    average_ratios_3 = np.nanmean([item['average_rationale_to_text_ratio'] for item in processed_data[key]['items'] if "average_rationale_to_text_ratio" in item])
    processed_data[key]['average_rationale_to_text_ratio'] = average_ratios_3
      
    correct_average_image_to_text_ratio = np.mean([
        item['average_image_to_text_ratio'] for item in processed_data[key]['items']
        if item['problem']['extracted_answer'] == item['problem']['answer']
    ])
    
    wrong_average_image_to_text_ratio = np.mean([
        item['average_image_to_text_ratio'] for item in processed_data[key]['items']
        if item['problem']['extracted_answer'] != item['problem']['answer']
    ])
    
    processed_data[key]['correct_average_image_to_text_ratio'] = correct_average_image_to_text_ratio
    processed_data[key]['wrong_average_image_to_text_ratio'] = wrong_average_image_to_text_ratio

    correct_average_image = np.mean([
        item['average_image'] for item in processed_data[key]['items']
        if item['problem']['extracted_answer'] == item['problem']['answer']
    ])
    
    wrong_average_image = np.mean([
        item['average_image'] for item in processed_data[key]['items']
        if item['problem']['extracted_answer'] != item['problem']['answer']
    ])
    
    processed_data[key]['correct_average_image'] = correct_average_image
    processed_data[key]['wrong_average_image'] = wrong_average_image
    
    processed_data[key]['correct_count'] = sum(1 for item in processed_data[key]['items'] if item['problem']['extracted_answer'] == item['problem']['answer'])
    processed_data[key]['wrong_count'] = sum(1 for item in processed_data[key]['items'] if item['problem']['extracted_answer'] != item['problem']['answer'])
    
  return processed_data
```

```python
import matplotlib.pyplot as plt

def plot_input_token_attention(processed_data, activations, idx):
  for key, data in processed_data.items():
    if activations[key] == 1:
      item = data['items'][idx]
      values = [item['average_image'], *item['average_text_tokens']]
  
      keys = ["avg_img", *item["main_input_text_tokens"]]
      print(key)
      # Plotting the bar graph
      plt.figure(figsize=(10, 6))
      plt.bar(keys, values, color='skyblue')
      plt.xlabel('Input Tokens')
      plt.ylabel(f'Attention Scores')
      plt.title(f'Input Token Average Attention')
      plt.xticks(rotation=45, ha='right')
      plt.tight_layout()
      plt.show()
```

```python
def plot_rationale_to_image_ratio(processed_data, activations):
    keys = list(processed_data.keys())
    filtered_keys = [key for key in keys if activations[key] and "average_rationale_to_image_ratio" in processed_data[key]]
    print(filtered_keys)
    
    average_ratios = [processed_data[key]["average_rationale_to_image_ratio"] for key in keys if activations[key] and "average_rationale_to_image_ratio" in processed_data[key]]
    print(average_ratios)

    # Plotting the bar graph
    plt.figure(figsize=(10, 6))
    plt.bar(filtered_keys, average_ratios, color='skyblue')
    plt.xlabel('Models')
    plt.ylabel('Average Rationale to Image Ratio')
    plt.title('How much did the answer attend to the rationale, compared to the input image?')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

def plot_rationale_to_text_ratio(processed_data, activations):
    keys = list(processed_data.keys())
    filtered_keys = [key for key in keys if activations[key] and "average_rationale_to_text_ratio" in processed_data[key]]
    print(filtered_keys)
    
    average_ratios = [processed_data[key]["average_rationale_to_text_ratio"] for key in keys if activations[key] and "average_rationale_to_text_ratio" in processed_data[key]]
    print(average_ratios)

    # Plotting the bar graph
    plt.figure(figsize=(10, 6))
    plt.bar(filtered_keys, average_ratios, color='skyblue')
    plt.xlabel('Models')
    plt.ylabel('Average Rationale to Text Ratio')
    plt.title('How much did the answer attend to the rationale, compared to the input text?')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt

def plot_average_image_to_text_ratios(processed_data, activations, type="all", key_="average_image_to_text_ratio"):
  keys = list(processed_data.keys())

  if type == "correct_wrong":
    correct_ratios = [processed_data[key]['correct_' + key_] for key in keys if activations[key]]
    wrong_ratios = [processed_data[key]['wrong_' + key_] for key in keys if activations[key]]

    # Print number of correct and wrong items
    correct_counts = [processed_data[key]['correct_count'] for key in keys if activations[key]]
    wrong_counts = [processed_data[key]['wrong_count'] for key in keys if activations[key]]
    for key, correct_count, wrong_count in zip(keys, correct_counts, wrong_counts):
        if activations[key]:
            print(f"Model: {key}, Correct: {correct_count}, Wrong: {wrong_count}")

    # Plotting the bar graph for both correct and wrong ratios
    x = [key for key in keys if activations[key]]
    x_indices = range(len(x))

    plt.figure(figsize=(10, 6))
    plt.bar(x_indices, correct_ratios, width=0.4, label='Correct', color='green', align='center')
    plt.bar(x_indices, wrong_ratios, width=0.4, label='Wrong', color='red', align='edge')
    plt.xlabel('Models')
    plt.ylabel('Average Image to Text Ratio')
    plt.title('Average Image to Text Ratio for Different File Names (Correct vs Wrong)')
    plt.xticks(ticks=x_indices, labels=x, rotation=45, ha='right')
    plt.legend()
    plt.tight_layout()
    plt.show()
  else:
    # Extract keys and their corresponding average_image_to_text_ratio values
    average_ratios = [processed_data[key][key_] for key in keys if activations[key]]

    print(average_ratios)

    # Plotting the bar graph
    plt.figure(figsize=(10, 6))
    plt.bar([key for key in keys if activations[key]], average_ratios, color='skyblue')
    plt.xlabel('Models')
    plt.ylabel(f'Average Image to Text Ratio ({type})')
    plt.title(f'Average Image to Text Ratio for Different File Names ({type})')
    plt.xticks(rotation=45, ha='right')
    
    # Set ylim between the min and max, plus a small buffer
    min_ratio = min(average_ratios)
    max_ratio = max(average_ratios)
    buffer = (max_ratio - min_ratio) * 0.1
    plt.ylim(min_ratio - buffer, max_ratio + buffer)
    
    plt.tight_layout()
    plt.show()

```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
import matplotlib.cm as cm
from matplotlib.colors import Normalize


def visualize_word_map(idx, processed_data, activations, key_="average_image_to_text_ratio"):
  # normalize ratios! 
  ratios = []
  for key, data in processed_data.items():
    if activations[key] == 1:
      items = data['items'][idx]["attention_results"]
      ratios.append([item[key_] for item in items])

  ratios_flat = [ratio for sublist in ratios for ratio in sublist]
  vmin = min(ratios_flat)
  vmax = max(ratios_flat)
  norm = lambda x: (x - vmin) / (vmax - vmin) if vmax != vmin else 0.5
  normalized_ratios_flat = [norm(ratio) for ratio in ratios_flat]
  
  final_ratios = []
  index = 0
  for sublist in ratios:
    final_ratios.append(normalized_ratios_flat[index:index + len(sublist)])
    index += len(sublist)
  
  # plot the graph!
  main_idx = 0
  cmap = cm.Oranges
  for key, data in processed_data.items():
    if activations[key] == 1:
      print(key)
      fig, ax = plt.subplots(figsize=(3, 2))
      item = data['items'][idx]

      x_coord = 0
      y_coord = 0.5
      for i, (output_token, normalized_value) in enumerate(zip(item["output_tokens"], final_ratios[main_idx])):
        color = cmap(normalized_value)
        ax.text(x_coord, y_coord, output_token, ha='left', va='center', fontsize=30, 
                bbox=dict(facecolor=color, edgecolor='none', boxstyle='round,pad=0.3'))
        x_coord += (len(output_token)/6)
        if (i + 1) % 10 == 0:
            x_coord = 0
            y_coord -= 0.5  # Shift down a little for every 5 tokens

      # Remove axes
      ax.axis('off')

      # Adjust layout
      plt.tight_layout()
      plt.show()
      main_idx += 1
```
